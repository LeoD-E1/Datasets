## Valores faltantes en un data set

Comprobar si faltan datos dentro de un data set y ser capaces de gestionar estos casos de forma adecuada es muy importante en el preprocesado de los datos. Si se dejan sin tratar, los valores que faltan pueden llvarnos a sacar una conclusion erronea.

```python
pd.isnull(titanic['body']) # Devuelve True o False segun corresponde de cada campo nulo de la columna indicada
```

```python
pd.notnull(titanic['body']) # Devuelve True o False segun corresponde de cada campo no nulo de la columna indicada
```

```python
pd.isnull(titanic['body']).values # Devuelve un array de datos que puede ocupar mas de una fila
```

```python
pd.isnull(titanic['body']).values.ravel() # Devuelve una unica fila contenido en un solo array
```

```python
pd.isnull(titanic['body']).values.ravel().sum() # Suma la cantidad de nulos que hay en la fila
```

### ¿Por que faltan valores?
Los valores que faltan en un data set pueden venir por dos razones:
* Extraccion de datos: incompatibilidad entre la base de datos y el lenguaje utilizado, por ejemplo con algun tipo de dato.
* Recoleccion de los datos: Puede que ya de pir si el formato de la recoleccion de datos sea imposible de extraer, por ejemplo en una encuesta, preguntarle a una persona por algo personal como su sueldo anual.

### ¿Que hacer cuando falten valores en un data set?

### Borrado de Valores faltantes

```python
titanic2 = titanic
```

```python
titanic2 = titanic2.dropna(axis=0, how="all") # Borra toda la fila si todos sus valores son NaN
titanic2.tail()
```

```python
titanic2 = titanic2.dropna(axis=0, how='any') # Borra toda la fila si al menos uno de sus valores es NaN
titanic2.tail() # En este caso, en el data set del titanic, todas las filas tenian al menos un dato Na
```

### Añadir o reemplazar valores faltantes por otros

```python
titanic3 = titanic
```

```python
# Reemplazara los valores faltantes con 0, sin importar si la columna es de tipo String
titanic3 = titanic3.fillna(0)
titanic3.tail()
```

```python
# Reemplazara los valores faltantes con "Desconocido" sin importar si la columna es de tipo numerico o floar
titanic3 = titanic3.fillna('Desconicido')
titanic3.tail()
```

### Sobreescritura con el tipo de dato de la columna

```python
titanic4 = titanic
```

```python
 # Reemplazara los valores faltantes por 0 solamente en la columna body
titanic4['body'] = titanic4['body'].fillna(0)
titanic4.tail()
```

```python
 # Reemplazara los valores faltantes por Desconocido solamente en la columna home.dest
titanic4['home.dest'] = titanic4['home.dest'].fillna('Desconocido')
titanic4.tail()
```

### Reemplazar valores faltantes por el promedio (solo en columnas numericas)

```python
titanic5 = titanic
```

```python
titanic5['age'] = titanic5['age'].fillna(titanic['age'].mean())
titanic5.tail()
```

### Reemplazar valores faltantes por el valor conocido mas cercano hacia arriba o hacia abajo

```python
titanic6 = titanic
```

```python
# Reemplaza los valores faltantes por el valor conocido mas cercano hacia ARRIBA
titanic6['age'] = titanic6['age'].fillna(method='ffill')
titanic6.tail()
```

```python
# Reemplaza los valores faltantes por el valor conocido mas cercano hacia ABAJO
titanic6['age'] = titanic6['age'].fillna(method='backfill')
titanic6.tail()
```

# Las variables Dummy
Las variables Dummy sirven para explicar valores cualitativos en un modelo de regresion. Los modelos de regresion intentan explicar una variable en funcion de otras. Por ejemplo, el salario de una persona en funcion de su edad, su educacion y la experiencia se pueden cuantificar en un numero de años. pero ¿Que ocurre ciando tenemos variables que no se pueden cuantificar? Por ejemplo el color de pelo, el pais de recidencia, o el sexo la solucion a estos problemas son las variables dummy. Suelen tomar valores binarios. Por ejemplo variable del sexo, es una variable cualitativa, para incluirla en in modelo de regresion hay que crear una variable dummy. A la dummy la llamaremos mujer. Si mujer = 1 (el individuo es mujer). Si mujer = 0 (el individuo es hombre).

```python
# Crea dos nuevas variables booleanas en base a la columna sex del 
# data set del titanic, como sex tiene 'male' y 'female', la dummy
# variable tendra dos variables 'sex_male', y  'sex_female'.
dummy_sex = pd.get_dummies(titanic['sex'], prefix='sex')
dummy_sex.head()
```

```python
# Una vez creada las dummy variables las tenemos que reemplazar dentro del data set original.
titanic = titanic.drop(['sex'], axis = 1)
titanic.head()
```

```python
# Agregar las columnas dummy al data set original
titanic = pd.concat([titanic, dummy_sex], axis = 1)
titanic.head()
```

```python
def createDummies(df_dev, var_name):
    dummy = pd.get_dummies(df_dev[var_name], prefix=var_name)
    df_dev = df_dev.drop([var_name], axis = 1)
    df_dev = pd.concat([df_dev, dummy], axis = 1)
    return df_dev
```
